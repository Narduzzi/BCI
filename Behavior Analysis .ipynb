```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import bokeh as bk
import os
```

```python
list_recordings = os.listdir("Recordings/")
list_recordings
```

```python
def createDFfor(folder):
    path = "Recordings/"+folder+"/"
    comparison = pd.read_csv(str(path)+str(folder)+"_comparison.csv")
    questionaire = pd.read_csv(str(path)+str(folder)+"_questionaire.csv",header=1)
    #Add folder for tracing
    questionaire["Folder"] = folder
    comparison["Folder"] = folder
    
    return comparison,questionaire
```

```python
def cleanColumns(columns):
    rep_col = []
    for c in columns:
        c = c.replace("\n"," ")
        rep_col.append(c)
    return rep_col
```

```python
comparisonDF = pd.DataFrame()
questionaireDF = pd.DataFrame()

for folder in list_recordings:
    comparison,questionaire = createDFfor(folder)
    comparison.columns = cleanColumns(comparison.columns)
    questionaire.columns = cleanColumns(questionaire.columns)
        
    comparisonDF = comparisonDF.append(comparison)
    questionaireDF = questionaireDF.append(questionaire)
```
